```python
covid_df = pd.read_csv('https://raw.githubusercontent.com/wcota/covid19br/master/cases-gps.csv')
```

```python
sp_df = covid_df[covid_df['name'].astype(str).str.contains('/SP')]
print(sp_df.shape)
sp_df.head()
```

```python
address = 'São Paulo, Brasil'

geolocator = Nominatim(user_agent="sp_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of São Paulo, Canada are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Manhattan using latitude and longitude values
map_sp = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(sp_df['lat'], sp_df['lon'], sp_df['name']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_sp)  
    
map_sp
```
